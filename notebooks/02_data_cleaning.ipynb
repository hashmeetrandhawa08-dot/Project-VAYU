```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/air_quality.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["AQI_Bucket"])
```

```python
numeric_cols = df.select_dtypes(include="number").columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
df.to_csv("../data/processed/clean_air_quality.csv", index=False)
```
